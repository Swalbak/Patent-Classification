```python
from transformers import AutoTokenizer
from torch.utils.data import Dataset
import torch

class cl_Dataset(Dataset):
    def __init__(self, data, attributes, max_token_len: int = 512, model_name: str = 'xlm-roberta-base'):
        self.data = data
        self.tokenizer = AutoTokenizer.from_pretrained(model_name)
        self.attributes = attributes
        self.max_token_len = max_token_len
    
    def __len__(self):
        return len(self.data)
    
    def __getitem__(self, index):
        item = self.data.iloc[index]
        abstract = item['Abstract (Original Language)']
        tokens = self.tokenizer.encode_plus(
            abstract,
            add_special_tokens=True,
            return_tensors='pt',
            truncation=True,
            padding='max_length',
            max_length=self.max_token_len,
            return_attention_mask=True
        )   
        label_tensor = torch.FloatTensor(item[self.attributes])
        return {'input_ids': tokens.input_ids.flatten(), 'attention_mask': tokens.attention_mask.flatten(), 'labels': label_tensor}

# cl_train_data = cl_Dataset(train_data, labels)
# cl_val_data = cl_Dataset(val_data, labels)

# print(cl_train_data.__getitem__(0))
# len(cl_train_data)
```

```python
from torch.utils.data import DataLoader

def read_data(data_path, max_token_len: int = 512, batch_size: int = 16, model_name: str = 'xlm-roberta-base'):
    data = pd.read_csv(data_path, index_col=False)

    #레이블 종류 구하기
    labels = set()
    IPC_list = list(data["IPC Current Full (4 Characters)"])
    Abstract_list = list(data["Abstract (Original Language)"])

    for i in range(len(IPC_list)):
        if pd.isna(IPC_list[i]):
            continue
        IPC_list[i] = list(map(lambda x: x.strip(), IPC_list[i].split('|')))
        labels.update(IPC_list[i])
    labels=list(labels)

    #레이블 추가
    for label in labels:
        data[label] = np.where(data['IPC Current Full (4 Characters)'].str.contains(label), 1, 0)

    #데이터 셋 나누기
    from sklearn.model_selection import train_test_split
    train_data, val_data = train_test_split(data, test_size=0.2, random_state=42)
    val_data, test_data = train_test_split(val_data, test_size=0.5, random_state=42)

    #데이터 셋 만들기
    train_dataset = cl_Dataset(data=train_data, attributes=labels)
    test_dataset = cl_Dataset(data=test_data, attributes=labels)
    val_dataset = cl_Dataset(data=val_data, attributes=labels)

    train_val_loaders = {
        'train': DataLoader(
            dataset=train_dataset,
            batch_size=batch_size,
            num_workers=4,
            shuffle=True
        ),
        'valid': DataLoader(
            dataset=val_dataset,
            batch_size=batch_size,
            num_workers=4,
            shuffle=False
        )
    }

    test_loaders = {
        'test': DataLoader(
            dataset=test_dataset,
            batch_size=batch_size,
            num_workers=4,
            shuffle=False
        )
    }
    return train_val_loaders, test_loaders
```

```python
import torch.nn as nn
from transformers import AutoModel, AutoConfig

class MyClassifier(nn.Module):
    def __init__(self, pretrained_model_name: str = 'xlm-roberta-base', num_classes: int = 600, dropout: float = 0.3 ):
        super().__init__()

        #config를 자동으로 불러옴
        config = AutoConfig.from_pretrained(pretrained_model_name, num_labels=num_classes)

        self.model = AutoModel.from_pretrained(pretrained_model_name, config=config)
        #은닉층, 출력층 연결
        self.classifier = nn.Linear(config.hidden_size, num_classes)
        self.dropout = nn.Dropout(dropout)
    
    def forward(self, features, attention_mask=None, head_mask=None):
        assert attention_mask is not None, 'attention mask is none'
        bert_output = self.model(input_ids=features, attention_mask=attention_mask, head_mask=head_mask)
        # we only need the hidden state here and don't need
        # transformer output, so index 0
        seq_output = bert_output[0] 
        # mean pooling, i.e. getting average representation of all tokens
        pooled_output = seq_output.mean(axis=1) 
        pooled_output = self.dropout(pooled_output) 
        scores = self.classifier(pooled_output) 

        return scores
```

```python
from utils import get_project_root
from catalyst.utils import prepare_cudnn
from catalyst.dl import CheckpointCallback, OptimizerCallback, SchedulerCallback, SupervisedRunner

train_val_loaders, test_loaders = read_data(train_path)
model = MyClassifier(pretrained_model_name='xlm-roberta-base')

criterion = torch.nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=float(1.5e-6))
scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer)

prepare_cudnn(deterministic=True)
```

```python
runner = SupervisedRunner(input_key=("input_ids", "attention_mask"))
```

```python
from catalyst.callbacks.metrics.accuracy import AccuracyCallback

runner.train(
    model=model,
    criterion=criterion,
    optimizer=optimizer,
    scheduler=scheduler,
    loaders=train_val_loaders,
    callbacks=[
        AccuracyCallback(num_classes=len(labels), input_key="logits", target_key="targets"),
        OptimizerCallback(accumulation_steps=4, metric_key="loss"),
        SchedulerCallback(loader_key="valid", metric_key="loss"),
        CheckpointCallback(logdir='logdir', loader_key="valid", metric_key="loss", minimize=True),
    ],
    logdir='logdir',
    num_epochs=2,
    verbose=True,
)

torch.cuda.empty_cache()

# getting validation metrics
metrics = runner.evaluate_loader(
    loader=train_val_loaders["valid"],
    callbacks=[AccuracyCallback(input_key="logits", target_key="targets")],
)
print(metrics)

# lastly, saving predicted scores for the test set
test_pred_scores = np.concatenate(
    [pred["logits"].detach().cpu().numpy() for pred in runner.predict_loader(loader=test_loaders["test"])]
)
np.savetxt(X=test_pred_scores, fname='result.txt')
```
