```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
sliced_num = 4
# sample_num = 400000
```

```python
df = pd.read_csv("/home/aailab/yjh/csv_data/patent_data/0913_patent_All_코드 파악 ML 용역 송부용_Final_UTF8.csv", index_col=0)
df
```

```python
# null값 확인
test_df = df[['Abstract (Original Language)', 'IPC - Current']].dropna()
test_df.isna().sum()
```

```python
# Abstract 길이 통계(이상치 확인)
(test_df['Abstract (Original Language)'].map(len)).describe()
```

```python
# 이상치 확인(길이가 긴 중복데이터)
test_df[test_df['Abstract (Original Language)'].map(len) > 20000]['Abstract (Original Language)']
```

```python
# 이상치 확인(길이가 1인 데이터 등 매우 짧은 데이터)
test_df[test_df['Abstract (Original Language)'].map(len) <= 100]['Abstract (Original Language)'].shape
```

```python
# 위의 숫자와 같음을 확인
test_df = test_df[(test_df['Abstract (Original Language)'].map(len) < 20000) & (test_df['Abstract (Original Language)'].map(len) > 100)]
test_df
```

```python
test_df = test_df.sample(frac=1, random_state=42)
test_df
```

```python
slice_num = sliced_num

labels = set()
IPC_Current = test_df['IPC - Current']
IPC_Current = IPC_Current.map(lambda x: x.split(' | '))
# sample들의 label
IPC_Nchar = []

for IPC_list in IPC_Current:
    sliced_IPC = list(map(lambda x: x[:slice_num], IPC_list))
    labels.update(sliced_IPC)
    IPC_Nchar.append(' | '.join(set(sliced_IPC)))
    
labels = list(labels)

len(IPC_Nchar), len(labels)
```

```python
test_df['IPC - Current'] = IPC_Nchar
test_df
```

```python
from sklearn.preprocessing import MultiLabelBinarizer

IPC_Current = test_df['IPC - Current']

mlb = MultiLabelBinarizer()
multilabel_list = mlb.fit_transform(IPC_Current.map(lambda x: x.split(' | ')).tolist())
multilabel_list.shape
```

```python
labels = mlb.classes_
```

```python
labels_df = pd.DataFrame(multilabel_list, columns=labels)
labels_df.head()
```

```python
test_num = 1
labels[multilabel_list[test_num] == 1], test_df.iloc[test_num]['IPC - Current']
```

```python
labels[labels_df.iloc[test_num] == 1], test_df.iloc[test_num]['IPC - Current']
```

```python
labels_df.index = test_df.index
```

```python
comp_df = pd.concat((test_df, labels_df), axis=1)
comp_df.head()
```

```python
test_df.iloc[0]['IPC - Current']
```

```python
comp_df.columns[np.where(comp_df.iloc[0] == 1)]
```

```python
save_path = f'/home/aailab/yjh/csv_data/patent_data/edited_data/BCELOSS/{sliced_num}sliced/'
comp_df.to_csv(save_path + 'comp_df.csv')
```

```python
# bert 학습을 위해 70만개를 기준으로 데이터 나누기
# 70만개는 버트 학습에 사용, 나머지를 test, val 데이터셋으로 나누기
train_csv = comp_df[:700000]
eval_csv = comp_df[700000:]
```

```python
sample_num_list = [20000, 200000, 400000, 'not']
```

```python
from sklearn.model_selection import train_test_split
import pickle

for sample_num in sample_num_list:
    train_df = None
    eval_df = None

    if sample_num != 'not':
        n = int(sample_num)
        train_df = train_csv.sample(n=n, random_state=42)
    else:
        n = 700000
        train_df = train_csv

    eval_df = eval_csv.sample(n=int(n*0.2), random_state=42)

    save_path = f'/home/aailab/yjh/csv_data/patent_data/edited_data/BCELOSS/{sliced_num}sliced/{sample_num}Sample/'
    # temp_df.to_csv(save_path + 'data.csv')

    val_df, test_df = train_test_split(eval_df, test_size=0.5, random_state=42)

    train_df.to_csv(save_path + 'train_data.csv')
    val_df.to_csv(save_path + 'val_data.csv')
    test_df.to_csv(save_path + 'test_data.csv')

    with open(save_path + "labels.pickle", "wb") as fw:
        pickle.dump(labels, fw)
```
